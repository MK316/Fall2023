### ðŸ”µ gTTS setting

```python
#@markdown ðŸ”„ Making functions: etts('text'), ktts('text')
%%capture
!pip install gTTS
from gtts import gTTS
from IPython.display import Audio

dialect = "us"
def etts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "en", tld = dialect,
                    slow = True)

  gtts_object.save("E-audio.mp3")
  return Audio("E-audio.mp3")

def ktts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "ko",
                    slow = False)

  gtts_object.save("K-audio.mp3")
  return Audio("K-audio.mp3")

def ftts(text):
  text_to_say = text

  gtts_object = gTTS(text = text_to_say,
                     lang = "fr",
                    slow = False)

  gtts_object.save("F-audio.mp3")
  return Audio("F-audio.mp3")
```

gTTS language options [visit](https://gtts.readthedocs.io/en/latest/module.html)

```python
#@markdown Transcribe what you hear:
text = "Jane kept repeating 'belts, little roses, n crayons.'"
etts(text)
```
